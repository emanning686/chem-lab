```python
import os
from rdkit import Chem
```

```python
with open('ligands/CR3.cif', 'r') as file:
    file_text = file.read()
print(file_text)
```

```python
def extractSmiles(file_text):
    file_lines = file_text.split('\n')
    smiles_line = next((item for item in file_lines if "SMILES_CANONICAL" in item), None)
    smiles = smiles_line.strip().split()[4].replace('"', "")
    return smiles

smiles = extractSmiles(file_text)

mol = Chem.MolFromSmiles(smiles)
mol
```

```python
with open('ligands/BAO.cif', 'r') as file:
    file_text = file.read()
smiles = extractSmiles(file_text)

mol = Chem.MolFromSmiles(smiles)
mol
```
