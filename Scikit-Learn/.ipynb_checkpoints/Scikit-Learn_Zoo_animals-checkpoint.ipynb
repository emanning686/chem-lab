# Scikit-Learn Zoo Animals Demo
## Goals
* __Separate data and train a machine learning model__ 
* __Predict an animals type based on attributes__

```python
from sklearn import svm
import pandas as pd
import math
```

```python
data = pd.read_csv("zoo.csv")
data.head(6) # class_type is a forein key to class.csv, which includes types: mammals, birds, etc
```

```python
from sklearn.model_selection import train_test_split

all_X = data.iloc[:, 1:17] # get all data except for the last column
all_y = data.iloc[:, 17] # get just the last column, class_type, because it is what we want to predict based on the other columns

X_train, X_test, y_train, y_test = train_test_split(all_X, all_y) # splits the data into 75% for training the model, and 25% for testing how well the model works
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# I found that LinearSVC makes for ~4% more accuracy, I have no idea what the difference is
#clf = svm.SVC()
clf = svm.LinearSVC() # choose the machine learning model
clf.fit(X_train, y_train) # train the data using the model
```

```python
clf.score(X_test, y_test) # test the test data and see how accurately it can predict the class_type based on the rest of the data
```

```python
clf.predict(X_test[10:15]) # predict specific rows
```
